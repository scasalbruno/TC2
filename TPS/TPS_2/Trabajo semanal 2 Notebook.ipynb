##### Se procede a hallar la transferencia T





* Se remarcan las primeras relaciones

$$ 
I_1 = \frac{V_i}{R_1} ~~~~~~~~  I_3= -\frac{V_o}{R_3} ~~~~~~~~~~~~ I_2 = I_1 - I_3 ~~~~~~~~~~~~ V_y = -V_o ~~~~~~~~~~~~ Z_1 = \frac{R_2}{1 + \$CR_2}
$$

* Calculamos el valor de nuestra corriente $I_2$ y nos das de resultado la siguiente relacion

$$ I_2 = \frac{V_i}{R_1} + \frac{V_o}{R_3}$$

* Nuestro valor de $I_2$ lo multiplicamos por $Z_1$ y nos el valor de $-V_x$

$$ V_x = -({\frac{V_i}{R_1}+\frac{V_o}{R_3}}).Z_1$$

* Si dividimos el valor $V_x$ con la resistencia $R_3$ nos da la corriente $I_4$

$$ \frac{V_x}{R_3} = I_4$$

* $I_4$ por el valor del capacitor C nos daria el valor de $V_Y$, que a su vez se sabe que $V_y = -V_o$

$$ V_y = I_4 .\frac{1}{\$C} ~~~~~~~~~~~~=> ~~~~~~~~~~~~~~~V_y = ({\frac{V_i}{R_3}+\frac{V_o}{R_3}}) . \frac{Z_1}{\$CR_3} = -V_o$$

* Reemplazando y despejando finalmente nos queda el siguiente resultado

$$T_{(\$)}=\frac{V_o}{V_i} = \frac{R_3}{R_1}\frac{\frac {1}{C^2R_3^2} }{\$^2 + \$ \frac{1}{CR_2}+\frac{1}{C^2R_3^2}}$$

* Normalizando nuestra red nos resulta se la siguiente forma:

$$ T_{(\$)} = K \frac{.W_o^2}{\$^2+\$\frac{W_o}{Q}+W_o^2}$$

* Donde nuestros valores son:
$$ K = \frac{R_3}{R_1} ~~~~~~ Q = \frac{R_2}{R_3} ~~~~~~ W_o = \frac{1}{CR_3}$$
<br>
* Con esto en mente podemos sacar la siguiente informacion:<br><br> 
<ul><li>$R_4$ es un seguidor, solo invierte el valor de $V_o$ en $V_y$</li>
    <li>$W_o$ queda definido por la relacion $\frac{1}{CR_3}$</li>
    <li>La relacion $\frac{R_3}{R_1}$ varia, la ganancia, pero debido a que nuestro $R_3$ pertenece a $W_o$ entonces nuestro grado de libertad en $K$ es $R_1$ </li>
    <li>El valor de $Q$ tambien pendende de nuestro valor de $R_3$ pero el que lo regula es el valor de $R_2$</li>
    
</ul>
   
    

* Realizamos una normalizacion de valores:
<ul><li> Definimos nuestro $W_o$ = 1 -> $\frac{1}{C}=R_3$ y le otorgamos el valor a $R_3$ = 1 -> $C$ = 1</li>
    <li> K = 10 -> $A_v$ = 20dB para $T_{(0)}$ -> $R_1$ = 0,1$R_3$ = 0,1</li>

##### Procedemos a sacar la sensibilidad del circuito

* Primero sacamos la sensibilidad de $W_o$ respecto al capacitor $C$ la cual es igual para $R_3$

$$ S_C^{W_o}= S_{R_3}^{W_o}= \frac{C}{W_o}\frac{∂W_o}{∂C}= C(R_3C)(-1)(R_3C)^{-2}R_3=-1 $$ 

* Realizamos la sensibilidad de $Q$ respecto al valor de $R_2$

$$ S_{R_2}^{Q}= \frac{R_2}{Q}\frac{∂Q}{∂R_2}=R_2 \frac{R_3}{R_2} \frac{1}{R_3} = 1$$

* Finalmente calculamos la sensibilidad de $Q$ respecto a $R_3$

$$ S_{R_3}^{Q}= \frac{R_3}{Q}\frac{∂Q}{∂R_3}=R_3 \frac{R_3}{R_2}(-1)\frac{R_2}{R_3^2} = -1$$

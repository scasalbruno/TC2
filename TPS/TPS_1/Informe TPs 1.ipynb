#####  Se realiza el cualculo de la transeferencia $H_{(\$)} =  \frac{V2}{V1}$ 

* Definimos las relaciones

$$
V_{x} = I_2 . R_3   ~~~~~~~~~~~~~~~~~~~~~~~~ V_i = I_2 . (\frac{1}{\$C}+R_3) ~~~~~~~~~~~~~~~~~~~~~~~~~ \frac{V_i - V_x}{R_1} = \frac{V_x-V_o}{R_2}
$$

<br>
- De la primrea relacion sacamos que $I_2 = \frac{V_x}{R_3}$ <br><br>
- Reemplazamos en la segunda y nos termina dando la siguiente relacion: $V_x = V_i . \frac{\$CR3}{\$CR3+1}$ <br><br>
- Finalmente reemplazamos en la ultima relacion: <br>
$$ \frac{V_i(1-\frac{\$R_3 C}{\$CR_3 + 1})}{R_1} = \frac{V_i \frac{\$R_3 C}{\$CR_3 + 1}-V_o}{R_2}$$ <br><br>
-Finalmente la relacion: $$\frac{V_o}{V_i} = \frac{\$ - \frac{R_2}{R_1.C.R_3}}{\$ + \frac{1}{CR_3}} $$

<br>

* Definimos el valor de: $W_o = \frac{1}{CR_3}$ <br><br>

- Reemplazando en nuestra transferencia nos queda: $$H_{(\$)}=\frac{V_o}{V_i}=\frac{\$-\frac{R_2}{R_1}W_o}{\$+W_o}$$ <br><br>

$Notar:$ Tanto $R_1$ como $R_2$ nos permite desplazar el valor de nuestro cero. <br><br>

* Observamos que nuestra red tiene un polo en $-W_o$ y tiene un cero $\frac{R_2}{R_1}W_o$. Con esta informacion podemos determinar que, para que nuestra red se comporte como un pasa todo la relacion debe ser igual entre $R_2$ y $R_1$
